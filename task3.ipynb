```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [200, 220, 250, 275, 300, 350],
    'Profit': [50, 60, 70, 80, 90, 100]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax1 = plt.subplots()

# Bar chart
ax1.bar(df['Month'], df['Sales'], color='b', label='Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis to plot the line chart
ax2 = ax1.twinx()
ax2.plot(df['Month'], df['Profit'], color='r', marker='o', label='Profit')
ax2.set_ylabel('Profit', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Title and legends
plt.title('Sales and Profit Over Time')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
